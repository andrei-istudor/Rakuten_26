```python
import os
import pandas as pd
from openai import OpenAI

key = os.environ.get("OPENAI_API_KEY")
# Initialize client (reads key from environment variable)
client = OpenAI(api_key=key)

df_proc = pd.read_csv("data/rakuten_processed.csv")
```

```python
def translate_to_english(text: str) -> str:
    """
    Translate arbitrary-language text to English using OpenAI.
    Handles None/NaN safely.
    """
    if text is None:
        return None
    if isinstance(text, float):  # catches NaN
        return None

    prompt = (
        "You are a translation engine. "
        "Translate the following text into **natural English**. "
        "Keep the meaning but do not add explanations.\n\n"
        f"Text:\n{text}"
    )

    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",  # Updated to a valid model
            messages=[{"role": "user", "content": prompt}],
            temperature=0.0,
        )
        translated = response.choices[0].message.content.strip()
        return translated
    except Exception as e:
        print("Error translating:", e)
        return None
```

```python
from tqdm import tqdm
import pandas as pd
import concurrent.futures

# Output file path
output_file = "data/rakuten_translated.csv"

# Filter rows that need translation (where text_english is None)
# We assume 'text_english' was initialized with None for non-English rows in previous cells
rows_to_translate = df[df['text_english'].isnull()].index
total_rows = len(rows_to_translate)
batch_size = 1000
max_workers = 20  # Adjust this based on your rate limits

print(f"Total rows to translate: {total_rows}")

# Process in batches
for i in range(0, total_rows, batch_size):
    # Identify the current batch indices
    batch_indices = rows_to_translate[i : i + batch_size]
    
    print(f"Processing batch {i // batch_size + 1} / {(total_rows // batch_size) + 1} (Rows {i} to {min(i + batch_size, total_rows)})")
    
    # Extract the texts to translate for this batch
    # We use .loc to ensure we get the rows by their original index
    texts_to_translate = df.loc[batch_indices, 'text_stripped'].tolist()
    
    # Translate each text in the batch using parallel processing
    translated_batch = []
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        # executor.map preserves the order of the input iterable
        results = list(tqdm(executor.map(translate_to_english, texts_to_translate), total=len(texts_to_translate), desc=f"Translating batch {i // batch_size + 1}"))
        translated_batch = results
    
    # Update the DataFrame
    df.loc[batch_indices, 'text_english'] = translated_batch
    
    # Save to CSV and reload as requested
    df.to_csv(output_file, index=False)
    # Reloading ensures we are working with the persisted state
    df = pd.read_csv(output_file)
    print(f"Saved progress to {output_file} and reloaded DataFrame.")

print("Translation complete.")
```

```python
df['token_len_translated'] = df['text_english'].astype(str).apply(lambda x: len(x.split()))
df.head(5)
```

```python
# add column text_english and token_len_translated to df_proc

KEYS = ["productid", "imageid"]
NEW_COLS = ["text_english", "token_len_translated"]

# 1) Drop old columns in df_proc (prevents _x/_y forever)
df_proc = df_proc.drop(columns=[c for c in NEW_COLS if c in df_proc.columns])

# 2) Build translation payload FROM THE TRANSLATION DF
# IMPORTANT: use ONE dataframe consistently (here: df_trans)
df_trans = df  # <-- rename explicitly if this is your translated df

payload = df_trans[KEYS + NEW_COLS].copy()

# 3) Enforce key uniqueness (fail fast)
if payload.duplicated(KEYS).any():
    dup = payload[payload.duplicated(KEYS, keep=False)][KEYS].head(10)
    raise ValueError(f"Duplicate keys in translation df for {KEYS}. Examples:\n{dup}")

if df_proc.duplicated(KEYS).any():
    dup = df_proc[df_proc.duplicated(KEYS, keep=False)][KEYS].head(10)
    raise ValueError(f"Duplicate keys in df_proc for {KEYS}. Examples:\n{dup}")

# 4) Merge safely (NO suffixes possible now)
df_proc = df_proc.merge(
    payload,
    on=KEYS,
    how="left",
    validate="one_to_one",
)


# 6) Save
df_proc.to_csv("data/rakuten_processed.csv", index=False)
```

# Separate Translation Approach

Translate designation and description separately, then combine them in two ways:
- `text_english_1`: Formatted as "Designation: ...; Description: ..."
- `text_english_2`: Simple concatenation with whitespace

```python
# Step 1: Translate designations
from tqdm import tqdm
import concurrent.futures

output_file = "data/rakuten_translated_separate.csv"

# Filter rows that need designation translation
rows_to_translate = df_separate[df_separate['designation_en'].isnull()].index
total_rows = len(rows_to_translate)
batch_size = 1000
max_workers = 20
print(f"Total rows to translate: {total_rows}")


print(f"Translating designations - Total rows: {total_rows}")

# Process in batches
for i in range(0, total_rows, batch_size):
    batch_indices = rows_to_translate[i : i + batch_size]
    
    print(f"Processing batch {i // batch_size + 1} / {(total_rows // batch_size) + 1} (Rows {i} to {min(i + batch_size, total_rows)})")
    
    # Extract designations to translate
    texts_to_translate = df_separate.loc[batch_indices, 'designation'].tolist()
    
    # Translate in parallel
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        results = list(tqdm(executor.map(translate_to_english, texts_to_translate), 
                          total=len(texts_to_translate), 
                          desc=f"Translating designations batch {i // batch_size + 1}"))
    
    # Update the DataFrame
    df_separate.loc[batch_indices, 'designation_en'] = results
    
    # Save and reload
    df_separate.to_csv(output_file, index=False)
    df_separate = pd.read_csv(output_file)
    print(f"Saved progress to {output_file}")

print("Designation translation complete.")
```

```python
# Step 2: Translate descriptions
from tqdm import tqdm
import concurrent.futures

output_file = "data/rakuten_translated_separate.csv"

# Reload to ensure we have the latest data

# Filter rows that need description translation
rows_to_translate = df_separate[df_separate['designation_en'].isnull()].index
total_rows = len(rows_to_translate)
batch_size = 1000
max_workers = 20

print(f"Translating descriptions - Total rows: {total_rows}")

# Process in batches
for i in range(0, total_rows, batch_size):
    batch_indices = rows_to_translate[i : i + batch_size]
    
    print(f"Processing batch {i // batch_size + 1} / {(total_rows // batch_size) + 1} (Rows {i} to {min(i + batch_size, total_rows)})")
    
    # Extract descriptions to translate
    texts_to_translate = df_separate.loc[batch_indices, 'description'].tolist()
    
    # Translate in parallel
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        results = list(tqdm(executor.map(translate_to_english, texts_to_translate), 
                          total=len(texts_to_translate), 
                          desc=f"Translating descriptions batch {i // batch_size + 1}"))
    
    # Update the DataFrame
    df_separate.loc[batch_indices, 'description_en'] = results
    
    # Save and reload
    df_separate.to_csv(output_file, index=False)
    df_separate = pd.read_csv(output_file)
    print(f"Saved progress to {output_file}")

print("Description translation complete.")
```

```python
# text_english_1 is designation_en + " " + description_en
df_separate['text_english_1'] = df_separate['designation_en'].astype(str) + " " + df_separate['description_en'].astype(str)
# text_english_2 is "Designation: " + designation_en + "; Description: " + description_en
df_separate['text_english_2'] = ("Designation: " + df_separate['designation_en'].astype(str) + "; Description: " + df_separate['description_en'].astype(str))

# and the lengths of deisgnation_en and description_en
df_separate['token_len_english_1'] = df_separate['text_english_1'].astype(str).apply(lambda x: len(x.split()))
df_separate['token_len_english_2'] = df_separate['text_english_2'].astype(str).apply(lambda x: len(x.split()))
```

```python
KEYS = ["productid", "imageid"]

NEW_COLS = [
    "designation_en",
    "description_en",
    "text_english_1",
    "text_english_2",
    "token_len_english_1",
    "token_len_english_2",
]

# 1) Remove old versions if they exist
df_proc = df_proc.drop(columns=[c for c in NEW_COLS if c in df_proc.columns])

# 2) Build payload
payload = df_separate[KEYS + NEW_COLS].copy()

# 3) Merge safely
df_proc = df_proc.merge(
    payload,
    on=KEYS,
    how="left",
    validate="one_to_one",
)

missing = df_proc["designation_en"].isna().sum()
if missing:
    raise ValueError(f"{missing} rows missing translations after merge.")
```
